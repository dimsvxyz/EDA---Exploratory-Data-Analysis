3. Cek Duplikat Data dan Nilai Unik

```python
import pandas as pd

try:
    df = pd.read_excel("Data Mental Health.xlsx")
except FileNotFoundError:
    df = pd.DataFrame()

total_duplikat = df.duplicated().sum()
print(f"Jumlah Data Duplikat : {total_duplikat}")
if total_duplikat > 0:
    print("Data Duplikat yang Ditemukan : ")
    print(df[df.duplicated()].head())
    print()
else:
    print("Tidak ada data duplikat ditemukan.\n")

print("Jumlah Nilai Unik Perkolom : ")
for kolom in df.columns:
    print(f"{kolom}: {df[kolom].nunique()}")
print()
```

4. Visualisasi Nilai Unik

    a. Tanpa filter (semua data)

```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    df = pd.read_excel("Data Mental Health.xlsx")
except FileNotFoundError:
    df = pd.DataFrame()

unique_counts = df.nunique()

print("Jumlah Nilai Unik Perkolom : ")
print(unique_counts, "\n")

plt.figure(figsize=(10, 5))
unique_counts.sort_values(ascending=False).plot(
    kind='bar',
    edgecolor='black'
)

plt.title("Jumlah Nilai Unik per Kolom", fontsize=14, pad=15)
plt.xlabel("Kolom", fontsize=12)
plt.ylabel("Jumlah Nilai Unik", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

    b. Dengan filter (menghilangkan ID)

Karena nilai unik ID sangat besar (banyak) sehingga data yang lain tidak terlihat maka kita filter dulu supaya yang lain bisa terlihat jelas.

```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    df = pd.read_excel("Data Mental Health.xlsx")
except FileNotFoundError:
    df = pd.DataFrame()

df_filtered = df.loc[:, df.nunique() < 1000]
unique_counts = df_filtered.nunique()

print("Jumlah Nilai Unik Perkolom : ")
print(unique_counts, "\n")

plt.figure(figsize=(10, 5))
unique_counts.sort_values(ascending=False).plot(
    kind='bar',
    edgecolor='black'
)

plt.title("Jumlah Nilai Unik per Kolom", fontsize=14, pad=15)
plt.xlabel("Kolom", fontsize=12)
plt.ylabel("Jumlah Nilai Unik", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

6. Replace Semua Null Value

```python
import pandas as pd

try:
    df = pd.read_excel("Data Mental Health.xlsx")
except FileNotFoundError:
    df = pd.DataFrame()

print("Jumlah Missing Values Sebelum Pembersihan : ")
print(df.isnull().sum(), "\n")

for col in df.columns:
    
    if df[col].dtype == 'O':  
        df[col].fillna("Tidak Diketahui", inplace=True)
    else: 
        df[col].fillna(0, inplace=True) 

print("Jumlah Missing Values Setelah Pembersihan : ")
print(df.isnull().sum())

df.to_excel("Data Mental Health Cleaned.xlsx", index=False)
```

9. Box Plot

    a. Tanpa filter (semua data)

```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    df = pd.read_excel("Data Mental Health.xlsx")
except FileNotFoundError:
    df = pd.DataFrame()

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Kolom numerik yang akan divisualisasikan : ")
print(numerical_cols, "\n")

plt.figure(figsize=(10, 6))
df.boxplot(column=numerical_cols)
plt.title("Box Plot untuk Kolom Numerik", fontsize=14, pad=15)
plt.ylabel("Nilai", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

    b. Tanpa filter (menghilangkan ID)

Sekali lagi ID akan dihilangkan karena jumlahnya yang sangat banyak dan mengganggu visualisasi data dalam bentuk box plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    df = pd.read_excel("Data Mental Health.xlsx")
except FileNotFoundError:
    df = pd.DataFrame()

df_filtered = df.loc[:, df.nunique() < 1000]
numerical_cols = df_filtered.select_dtypes(include=['int64', 'float64']).columns.tolist()

print("Kolom numerik yang akan divisualisasikan : ")
print(numerical_cols, "\n")

plt.figure(figsize=(10, 6))
df.boxplot(column=numerical_cols)
plt.title("Box Plot untuk Kolom Numerik", fontsize=14, pad=15)
plt.ylabel("Nilai", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

    C. Satu satu

Karena masih ada beberapa box plot dari keseluruhan data yang masih belum terlihat, bisa juga divisualisasikan satu satu.

```python
import pandas as pd
import matplotlib.pyplot as plt

try:
    df = pd.read_excel("Data Mental Health.xlsx")
except FileNotFoundError:
    df = pd.DataFrame()

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

if not numerical_cols:
    print("Bukan kolom numerik")
else:
    print("Kolom numerik yang akan divisualisasikan : ")
    print(numerical_cols, "\n")

    for col in numerical_cols:
        plt.figure(figsize=(6, 4))
        plt.boxplot(df[col].dropna(), vert=True, patch_artist=True)
        plt.title(f"📦 Box Plot: {col}", fontsize=14, pad=10)
        plt.ylabel("Nilai", fontsize=12)
        plt.grid(axis='y', linestyle='--', alpha=0.6)
        plt.tight_layout()
        plt.show()
```

10. Korelasi

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

try:
    df = pd.read_excel("Data Mental Health.xlsx")
except FileNotFoundError:
    df = pd.DataFrame()

numerical_df = df.select_dtypes(include=['int64', 'float64'])

print("Kolom numerik untuk korelasi : ")
print(numerical_df.columns.tolist(), "\n")

correlation = numerical_df.corr()

print("Matriks Korelasi : \n")
print(correlation, "\n")

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap="coolwarm", center=0, fmt=".2f")
plt.title("Korelasi antar Variabel Numerik", fontsize=14, pad=12)
plt.tight_layout()
plt.show()
```
