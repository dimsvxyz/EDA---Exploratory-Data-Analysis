```python
import pandas as pd

file_path = "Data Mental Health.xlsx"
df = pd.read_excel(file_path)

print("Kolom tersedia:", df.columns.tolist())
print("Nilai unik Gender:", df['Gender'].unique())

df_female = df[df['Gender'].str.lower() == 'female']
print(f"Jumlah total data: {len(df)}")
print(f"Jumlah data Female: {len(df_female)}")
print(f"Persentase Female: {len(df_female)/len(df)*100:.2f}%")

df_female.head(10)
```

```python
df_age = df[df['Age'] < 300]

print("Jumlah Total Data:", len(df))
print("Jumlah Data dengan Usia < 300:", len(df_age))
print("Persentase (%):", round(len(df_age)/len(df)*100, 2))

df_age.head(10)
```
